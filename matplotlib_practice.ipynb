```python
#Import Dependencies. 

import matplotlib.pyplot as plt 
```

```python
# Set the x-axis to a list of strings for each month. 
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug" , "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fara in US Dollars accumulated for each month. 
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

```

```python
# Create the plot. 
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plot()

fig, ax = plt.subplots()

ax.plot(x_axis, y_axis)
```

```python
# Create a plot with object oriented interface 

fig = plt.figure()

ax = fig.add_subplot()

ax.plot(x_axis, y_axis)
```

```python
# Create the plot using ax.plot() method. 

ax = plt.axes()

ax.plot(x_axis, y_axis)
```

```python
# Create the plot. 

plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend. 
plt.plot(x_axis, y_axis, label = "Boston")

# Create labels for the x and y axes. 
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45. 
plt.ylim(0,45)

# Create a title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
# Create the plot. 
plt.plot(x_axis, y_axis, marker = "*", color = "blue", linewidth = 2, label = "Boston")

# Create labels for the x and y axes. 
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45. 
plt.ylim(0,45)

# Create a title. 
plt.title("PyBer Fare By Month")

# Add a grid. 
plt.grid()

# Add the legend. 
plt.legend()

```

```python
# Using the Object-Oriented approach, skill drill 

# Change the line color to green. 
fig,ax = plt.subplots(figsize = (20,20))

ax.plot(x_axis, y_axis, marker = "*", color = "green", linewidth = 2, label = "Boston")

ax.set_title("PyBer Fare By Month")
ax.legend()
ax.set_xlabel("Time")
ax.set_ylabel("Fare($)")
ax.grid()
```
